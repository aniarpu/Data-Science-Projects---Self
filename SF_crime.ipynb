```python
# get the year in which the crime occurred

train['year'] = train['Dates'].map(lambda date: date.split('-')[0])
```

```python
# similarly get the month and hour inwhich the crime occurred

train['month'] = train['Dates'].map(lambda date: date.split('-')[1])
train['hour'] = train['Dates'].map(lambda date: date.split('-')[2].split(' ')[1].split(':')[0])
```

```python
hour = train['hour'].values
```

```python
# convert the hour 00 into 24

for i in range(np.size(hour)):
    if(hour[i] == '00'):
        hour[i] = '24'
```

```python
train['hour'] = hour
```

```python
# the dates are now useless since we have the year, month and hour of the crime

train.drop(['Dates'],axis = 1,inplace = True)
```

```python
# create a formula for dmatrice to use and create new training andtest dataframes

formula = 'Category ~ C(DayOfWeek) + X + Y + C(year) + C(month) + C(hour) + C(PdDistrict) '
```

```python
y_train,x_train = dmatrices(formula,train,return_type = "dataframe")
```

```python
y_test,x_test = dmatrices(formula,test,return_type = "dataframe")
```

```python
np.shape(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV
```

```python
t1 = train.sample(frac=1).reset_index(drop=True)
```

```python
y_1,x_1 = dmatrices(formula,t1[t1.index < 300],return_type = 'dataframe')
```

```python
np.shape(x_1)
```

```python
# Now to do feature selection because we can't have 64 features
```

```python
# We use Random forest Classifier to train the data
clf_1 = RandomForestClassifier()
```

```python
clf_1.fit(x_1,y_1)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
np.shape(x_1)
```

```python
# We take out all the features which do not affect the predicted outputs or which have very less weights in the formula

model = SelectFromModel(clf_1,prefit = True)
```

```python
x_2 = model.transform(x_1)
```

```python
np.shape(x_2)
```

```python
# hyper parameter tuning for finding the most optimum model using GridSearch cross validation

parameter_grid = {
                 'max_depth' : [4, 6, 8],
                 'n_estimators': [50, 10],
                 'max_features': ['sqrt', 'auto', 'log2'],
                 'min_samples_split': [2, 5, 10],
                 'min_samples_leaf': [1, 4, 10],
                 'bootstrap': [True, False],
                 }
forest = RandomForestClassifier()
cross_validation = StratifiedKFold(t1[t1.index < 300].Category, n_folds=5)
grid_search = GridSearchCV(forest,
                               scoring='accuracy',
                               param_grid=parameter_grid,
                               cv=cross_validation)

grid_search.fit(x_2,y_1 )
model = grid_search
parameters = grid_search.best_params_
```

```python
# these are the parameters for the optimum model
parameters
```

```python
clf = RandomForestClassifier(bootstrap = False,max_depth = 8,max_features = 'auto',min_samples_leaf = 1,min_samples_split = 2,n_estimators = 10)
```

```python
# train the new model with optimum parameters

clf.fit(x_train,y_train)
```

```python
# classify the test data with the help of the trained model

y_predicted = clf.predict(x_test)
```
